Disini saya akan mencoba menghitung persentase kemungkinan player akan mendapatkan item atau karakter bintang 5 sebelum terjamin dapat di undian ke 90.


**1. Parameter**



```python
import matplotlib.pyplot as plt
import numpy as np
a = 17
m = 61
z0 = 3
n = 89
```

ùëé: Konstanta untuk menentukan skala perhitungan.

ùëö: Modulus untuk memastikan hasil tetap dalam batas tertentu (0 hingga ùëö ‚àí 1).

z0: nilai awal penghitungan

n: jumlah iterasi yang diinginkan (saya gunakan 90)

Library matplotlib dan numpy digunakan untuk menampilkan grafik di akhir

**2. Inisialisasi Data (Array)**

```python
results = []
```

results = [] adalah list yang digunakan untuk menyimpan data hasil perhitungan algoritma Random Number Generator (RNG)

**3. Algoritma RNG**

```python
z = z0
for i in range(1, n+1):
 z_prev = z
 z = (a * z) % m
 u = z/m
 percentage = round(u * 100, 2)
 rumus = f"(17 * {z_prev}) mod {m}"
 results.append([i, z_prev, rumus, z, round(u, 3), f"{percentage}%"])
results.append([90, "PASTI DAPAT", "PASTI DAPAT", "PASTI DAPAT", 1, "100%"])
```

z = z0 Menetapkan nilai awal z0, yang merupakan seed (benih) untuk menghasilkan bilangan acak.

Perulangan for berjalan sebanyak n kali, yaitu dari 1 hingga n (dalam hal ini, 89).

Nilai z saat ini disimpan ke dalam z_prev. Ini digunakan untuk mencetak bilangan acak sebelumnya di tabel hasil.

z = (a * z) % m adalah Rumus LCG digunakan untuk menghitung z baru

u = z / m Nilai u dihitung sebagai rasio z terhadap m, sehingga menghasilkan nilai dalam interval [0, 1], yang menyerupai bilangan acak seragam.

Nilai u dikalikan dengan 100 untuk mengubahnya menjadi persentase.

Hasil perhitungan untuk iterasi saat ini ditambahkan ke dalam list results

Iterasi ke-90 ditambahkan secara manual dengan nilai-nilai khusus:

90, "PASTI DAPAT", "PASTI DAPAT", "PASTI DAPAT", 1, "100%"

karena dalam sistem wish genshin impact ini, pada wish ke 90 pasti akan mendapatkan item/karakter bintang 5 yang tersedia

**4. Output dalam bentuk tabel**

```python
print(f"{'1':<3} | {'zi-1(Bilangan Sebelumnya)':<28} | {'rumus zi = (17,zi-1) mod 61':<32} | {'z (Bilangan Acak)':<20} | {'ui = zi/61 (Bilangan Acak Seragam)':<35} | {'% (Persentase)':<10}")
print("_" * 150)

for row in results:
 print(f"{row[0]:<3} | {row[1]:<28} | {row[2]:<32} | {row[3]:<20} | {row[4]:<35} | {row[5]:<10}")
```

**5. Output Dalam Bentuk Grafik**

```python
iterasi = [row[0] for row in results]
value = [row[4] for row in results]

plt.figure(figsize=(10, 6))
plt.plot(iterasi, value, marker='o', linestyle='-', color='b', label='Nilai Ui')
plt.axhline(y=1, color='r', linestyle='--', label='Ui = 1 (Guaranteed)')
plt.title("Grafik Bilangan Acak Seragam (Ui) dari LCG")
plt.xlabel("Iterasi")
plt.ylabel("Nilai Ui")
plt.grid(True)
plt.legend()
plt.show()
```

Bagian grafik ini membuat plot untuk nilai Ui yang dihitung dari Linear Congruential Generator (LCG) selama 90 iterasi:

1. Data untuk Grafik: Mengambil iterasi dan nilai Ui dari hasil perhitungan.
2. Plotting: plt.plot() menggambar grafik dengan iterasi di sumbu X dan nilai Ui di sumbu Y, menggunakan titik marker dan garis biru.
3. Garis Horizontal: Menambahkan garis merah di Ui = 1 pada iterasi ke-90 untuk menandakan bahwa nilai Ui dijamin menjadi 1.
4. Judul dan Label: Menambahkan judul, label sumbu X (Iterasi), dan label sumbu Y (Nilai Ui).
5. Grid dan Legenda: Menambahkan grid untuk kemudahan membaca dan legenda untuk membedakan dua garis.

Grafik ini menunjukkan perubahan nilai Ui selama 90 iterasi LCG.
